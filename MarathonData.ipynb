```python
#Difference in speed for the 50k, 50mi for male to female
```

```python
#Finding month in which each race was ran
```

```python
ReorderedDF['Race_Month'] = ReorderedDF['Race_Day'].str.split('.').str.get(1).astype(int)
```

```python
ReorderedDF.head()
```

```python
#Finding season in which each race was ran
```

```python
ReorderedDF['Race_Season'] = ReorderedDF['Race_Month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
ReorderedDF.head(25)
```

```python
#Finding average speed per season
```

```python
ReorderedDF.groupby('Race_Season')['Athlete_Average_Speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
