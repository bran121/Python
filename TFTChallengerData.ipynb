```python
# What are the highest win rate Little Legends in the game?
ax7 = new_challenger_data.groupby('companion_species')['placement'].agg(['mean', 'count']).sort_values(by='mean', ascending=True)
top_20 = ax7.head(20)

# Specify colors for the first 5 bars individually
colors = ['red', 'green', 'blue', 'orange', 'purple']
default_color = 'gray'  # Default color for bars after the first 5

# Create a list of colors for all bars
all_colors = colors + [default_color] * (len(top_20) - len(colors))

# Plotting
top_20.plot(kind='bar', y='mean', title='Highest Win Rate Little Legends', xlabel='Little Legend', ylabel='Placement', color=all_colors)
plt.show()
```
